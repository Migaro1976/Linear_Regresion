### EJEMPLO USO FUNCION POLYNOMIAL

* Primero importamos la funcion polynomial como pol.
* Despues con la opción thetas tenemos los diferentes parámetros de la función, es decir, para una función cuadrática será pol.thetas(2) 
* Si queremos ver algún parámetro específico indicamos el indice,  por ejemplo para theta2: pol.thetas(2)[1]

```python
#FUNCION POLINOMICA DE ORDEN 1 (llamamos a la funcion por thetas(1))
import polynomial as pol

print(f'Y = {pol.thetas(1)[0]:.2f} {pol.thetas(1)[1]:+.2f}*x')

import matplotlib.pyplot as plt
fig, bx = plt.subplots(figsize=(12,8))
bx.plot(pol.X,pol.thetas(1)[0]+pol.thetas(1)[1]*pol.X,'g-')
bx.scatter(pol.X, pol.y)
bx.grid(True)
```

```python
#FUNCION POLINOMICA DE ORDEN 2 (llamamos a la funcion por thetas(2))
import polynomial as pol

print(f'Y = {pol.thetas(2)[0]:.2f} {pol.thetas(2)[1]:+.2f}*x {pol.thetas(2)[2]:+.2f}*x\u00B2')

import matplotlib.pyplot as plt
fig, bx = plt.subplots(figsize=(12,8))
bx.plot(pol.X,pol.thetas(2)[0]+pol.thetas(2)[1]*pol.X+pol.thetas(2)[2]*pol.X**2,'g-')
bx.scatter(pol.X, pol.y)
bx.grid(True)
```

```python
#FUNCION POLINOMICA DE ORDEN 3 (llamamos a la funcion por thetas(3))
import polynomial as pol

print(f'Y = {pol.thetas(3)[0]:.2f} {pol.thetas(3)[1]:+.2f}*x {pol.thetas(3)[2]:+.2f}*x\u00B2 {pol.thetas(3)[3]:+.2f}*x\u00B3')

import matplotlib.pyplot as plt
fig, bx = plt.subplots(figsize=(12,8))
bx.plot(pol.X,pol.thetas(3)[0]+pol.thetas(3)[1]*pol.X+pol.thetas(3)[2]*pol.X**2+pol.thetas(3)[3]*pol.X**3,'g-')
bx.scatter(pol.X, pol.y)
bx.grid(True)
```
